1- Abrindo arquivo

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df_train = pd.read_csv("C:/Utilidades/Python/Aulas/df_train.csv")
df_train
```

2- Separando as colunas nas variaveis Y e X

```python
X = df_train.loc[:,"Pclass":"Embarked"]
Y = df_train.loc[:,"Survived"]
print(f"Valores de X:\n{X} \n\nValores de Y\n{Y}")
```

3- Separando os grupos de Treino e Teste

```python
X_train, X_teste, Y_train, Y_teste = train_test_split(X, Y)

X_train
```

4- Colocando KNN em uma variÃ¡vel

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

5- Fazendo uma prediÃ§Ã£o

```python
knn.fit(X_train, Y_train)
y_pred = knn.predict(X_teste)
print("PrecissÃ£o do algoritimo, com k=5:", accuracy_score(Y_teste, y_pred)*100)
```
