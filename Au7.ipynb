1- Criando uma matriz aleatória e exibindo-a

```python
import pandas as pd
import numpy as np

matriz_1 = np.random.rand(10, 4)
col = ["age", "mean_N1", "mean_N2", "IRA"]

df_1 = pd.DataFrame(data=matriz_1, columns=col)
df_1.head(3)
```

2- Adicionando nova coluna (mean_Final)

```python
N=0

while N<=9:
    Res = (df_1.loc[N,"mean_N1"] + df_1.loc[N,"mean_N2"])/2
    df_1.loc[N,"mean_Final"] = Res
    N=N+1

df_1
```

3- Descrição estatística do Data Frame

```python
df_1.describe()
```

4- Criando outro Data Frame para ser concatenado com o anterior

```python
df_2 = pd.DataFrame({"id": [21312, 41234, 12323, 25324, 23423, 98345, 34234, 98289, 23487, 23453],
                     "name": ["Carla","João","Maria","Alberto","Gustavo","Amanda","Luiz","Davyd","Elise","Evylin"]})

df_new = df_1.join(df_2)
df_new
```
