1- Criando um sinal 

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,1,1000)
freq = 20
Sinal = np.sin(2*np.pi*freq*t)

plt.figure(figsize=(10,5))
plt.plot(t, Sinal, label = "Sinal original")
plt.legend()
plt.grid(True)
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.title("Grafico do sinal")
plt.show()
```

2- Aplicando Fourier

```python
trans = np.fft.fft(Sinal)
f_freq = np.fft.fftfreq(len(t),t[1] - t[0])

plt.figure(figsize=(10,5))
plt.plot(f_freq, np.abs(trans))
plt.xlabel("Frequncia (Hz)")
plt.ylabel("Magnitude")
plt.title("Transformada de Fourier")
plt.grid(True)
plt.show
```

3- Filtrando o Sinal

```python
ruido = 1 * np.random.randn(len(t))
sinal_ruim = Sinal + ruido

c_freq = 21
trans_ruim = np.fft.fft(sinal_ruim)
trans_ruim[c_freq < f_freq] = 0
sinal_filtrado = np.fft.ifft(trans_ruim)

plt.figure(figsize=(10,5))
plt.plot(Sinal,label= "Sinal original")
plt.plot(sinal_ruim,"r",label="Sinal com ruido")
plt.plot(sinal_filtrado,"g",label="Sinal filtrado")
plt.legend()
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

4- Fazendo a Anlise Espectral

```python
espec = np.abs(trans)

plt.figure(figsize=(10,5))
plt.plot(f_freq, espec)
plt.xlabel("Frequncia (Hz)")
plt.ylabel("Magnitude")
plt.title("Anlise Espectral")
plt.grid(True)
plt.show()
```
