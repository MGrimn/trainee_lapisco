1- Removendo ruidos da imagem 1

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

#Acredito que ser necessrio trocar o pathing para o arquivo
#Imagem usada "Gustavo Davyd - Captura de tela 2023-07-22 080126.png"

img = cv2.imread("C:/Users/Usuario/Downloads/Img_1.png")

img_tret = cv2.medianBlur(img,5)

plt.figure(figsize=(8,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis(False)
plt.title("Antes")
plt.show()

plt.figure(figsize=(8,6))
plt.imshow(cv2.cvtColor(img_tret, cv2.COLOR_BGR2RGB))
plt.title("Depois")
plt.axis(False)
plt.title("Depois")
plt.show()
```

2- Tratando imagem 2

```python
#Imagem usada "Gustavo Davyd - images.jpg"

img_2 = cv2.imread("C:/Users/Usuario/Downloads/Img_2.jpg")

img_2_grey = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)

img_2_tret = cv2.medianBlur(img_2_grey,3)

img_2_norm = cv2.normalize(img_2_tret, True, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

plt.figure(figsize=(8,6))
plt.imshow(img_2)
plt.axis(False)
plt.title("Antes")
plt.show()

plt.figure(figsize=(8,6))
plt.imshow(img_2_norm, cmap= "gray")
plt.axis(False)
plt.title("Depois")
plt.show()
```

3- Normalizando a imagem 3

```python
#Imagem usada "Gustavo Davyd - download.jpg"

img_3 = plt.imread("C:/Users/Usuario/Downloads/Img_3.jpg")

img_3_norm = cv2.normalize(img_3, True, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

plt.figure(figsize=(5,5))
plt.imshow(img_3)
plt.axis(False)
plt.title("Antes")
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(img_3_norm, cmap= "gray")
plt.axis(False)
plt.title("Depois")
plt.show()
```
